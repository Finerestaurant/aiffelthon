```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.random.randn(19, 19)
fig1, ax1 = plt.subplots()
im = ax1.imshow(a, aspect='auto', origin='lower', interpolation='none')
fig1.colorbar(im)
fig1.savefig('1.png')


fig2, ax2 = plt.subplots()
im1 = ax2.imshow(a, aspect='auto', origin='lower', interpolation='none')
fig2.colorbar(im1)
fig2.savefig('2.png')
```
